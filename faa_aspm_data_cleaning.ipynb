```python
import pandas as pd

file_path = "APM-Report-16896.xls"

# Read all tables from the HTML report
tables = pd.read_html(file_path, flavor="lxml")

print(len(tables))  # see how many tables
first_sheet = tables[0]  # usually first table is what you want

first_sheet.head()
```

```python
print(first_sheet.columns)
print(first_sheet.head(10))
```

```python
df = first_sheet.copy()

# Flatten multiindex
df.columns = [col[-1] if isinstance(col, tuple) else col for col in df.columns]
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
df
```

```python
df = df.rename(columns={
    'facility': 'airport',
    'calendar_year': 'year',
    'delayed_gate_arrivals': 'delayed_arrivals',
    'arrivals_for_metric_computation': 'arrivals',
    '%_delayed_gate_arrivals': 'delay_rate',
    'average_minutes_of_delay_per_delayed_gate_arrival': 'avg_delay'
})
df
```

```python
# keeping only useful columns
df = df[[
    'airport',
    'year',
    'arrivals',
    'delayed_arrivals',
    'delay_rate',
    'avg_delay'
]]
df
```

```python
# fixing data types
df['year'] = pd.to_numeric(df['year'], errors='coerce')

for col in ['arrivals', 'delayed_arrivals', 'delay_rate', 'avg_delay']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df
```

```python
df['delay_rate_calc'] = df['delayed_arrivals'] / df['arrivals']
```

```python
airport_features = df.groupby('airport').agg({
    'delay_rate': 'mean',
    'avg_delay': 'mean'
}).rename(columns={
    'delay_rate': 'avg_delay_rate',
    'avg_delay': 'avg_delay_minutes'
}).reset_index()
# Convert percentage to decimal
airport_features['avg_delay_rate'] = airport_features['avg_delay_rate'] / 100
```

```python
volatility = df.groupby('airport')['delay_rate'].std().reset_index()
volatility = volatility.rename(columns={'delay_rate': 'delay_volatility'})

airport_features = airport_features.merge(volatility, on='airport')
```

```python
airport_features['gdp_proxy_score'] = airport_features['avg_delay_rate']
airport_features.head()
```

```python
output_csv = 'airport_features.csv'
airport_features.to_csv(output_csv, index=False)
print(f"Exported {len(airport_features)} rows to {output_csv}")
```
